4) PROGRAM TO CONSTRUCT A TREE AND DISPLAY ITS ARCHITECHTURE

```python
#importing decision tree classifier 
from sklearn.tree import DecisionTreeClassifier 
clf = DecisionTreeClassifier()
```

```python
iris.feature_names
```

```python
x_train=df_train[iris.feature_names]
x_test=df_test[iris.feature_names]
```

```python
y_train=df_train['target']
y_test=df_test['target']
```

```python

#training decision tree classifier
clf.fit(x_train,y_train)
print("Accuracy of training data is : ", clf.score(x_train, y_train))

```

```python
#prediction on test data
y_pred=clf.predict(x_test)
```

```python
#To display the decision tree in command shell
from sklearn.tree import export_text
from sklearn import tree
from matplotlib import pyplot as plt

text_representation= tree.export_text(clf)
print(text_representation)

```

8) PROGRAM TO GET SCORE FROM 100 ITERATIONS

```python
# initializes an empty list called accuracy_scores,
# used to store the accuracy scores calculated for different iterations.
accuracy_scores = []


for _ in range(100):  # Change 5 to 100 for 100 iterations
    # At each iteration, there is a fresh split
    df_train, df_test = train_test_split(df, test_size=0.3)

    x_train = df_train[iris.feature_names]
    x_test = df_test[iris.feature_names]
    y_train = df_train['target']
    y_test = df_test['target']

    clf = DecisionTreeClassifier()
    clf.fit(x_train, y_train)
    y_pred = clf.predict(x_test)
    accuracy_scores.append(round(accuracy_score(y_test, y_pred), 3))

accuracy_scores = pd.Series(accuracy_scores)

print('Accuracy Scores for 100 iterations are as follows:')
print(accuracy_scores)
```

9) PROGRAM TO SHOW SHUFFLE SPLIT

'ShuffleSplit' :cross-validator is created with specific settings:
n_splits=100: This indicates that the dataset will be split into 100 different train-test pairs.
test_size=0.3: Each split will have a test set size of 30% of the total dataset.

```python
from sklearn.model_selection import ShuffleSplit


print('PROGRAM TO show shuffle split')
accuracy_scores=[]
rs=ShuffleSplit(n_splits=100,test_size=0.3)
for train_index,test_index in rs.split(df):
 x_train=df.loc[train_index,iris.feature_names]
 x_test=df.loc[test_index,iris.feature_names]
 y_train=df.loc[train_index,'target']
 y_test=df.loc[test_index,'target']

 clf=DecisionTreeClassifier()

 clf.fit(x_train,y_train)
 y_pred=clf.predict(x_test)

 accuracy_scores.append(round(accuracy_score(y_test,y_pred),3))

# list of accuracy scores is converted into a pandas Series object for further analysis and printing.
accuracy_scores=pd.Series(accuracy_scores)
print('Accuracy Scores for 100 iterations are as follows')
print(accuracy_scores)
```

10) Program to show Cross Validation

The cross_validate function is called to perform cross-validation. It takes the following arguments:
clf: The classifier to be evaluated (Decision Tree Classifier).
x: The feature data.
y: The target variable.
cv: The cross-validator object (rs) that defines how the data will be split.
scoring='accuracy': The scoring metric to be used, which is accuracy in this case.

```python
print('This program shows Cross Validation')
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_validate

clf=DecisionTreeClassifier()
rs=ShuffleSplit(n_splits=100,test_size=0.3)

x=df[iris.feature_names]
y=df['target']

#Accuracy for different iterations
print('Accuracy for different iterations')
cv_results=cross_validate(
 clf,x,y,cv=rs,scoring='accuracy'
 ) 
# list of accuracy scores is converted into a pandas Series object for further analysis and printing.
accuracy_scores=pd.Series(cv_results['test_score'])
print(accuracy_scores)
```

```python
#Precision for different iterations
print('Precision for different iterations')
cv_results=cross_validate(
 clf,x,y,cv=rs,scoring='precision_macro'
 ) 
precision_scores=pd.Series(cv_results['test_score'])
print(precision_scores)
```

```python
#Recall for different iterations
print('Recall for different iterations')
cv_results=cross_validate(
 clf,x,y,cv=rs,scoring='recall_macro'
 ) 
recall_scores=pd.Series(cv_results['test_score'])
print(recall_scores)
```
